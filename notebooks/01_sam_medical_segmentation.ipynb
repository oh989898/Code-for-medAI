## 1. 数据准备

### 1.1 快速可运行示例（玩具数据）
**做什么**：生成一个简单的“圆形”二值掩码，用它模拟肿瘤区域。
**为什么**：让你在没有真实数据的情况下，也能快速跑通流程。
**结果**：得到一张图像和对应的分割掩码。

````python
import numpy as np
import matplotlib.pyplot as plt
from skimage.draw import disk

image = np.zeros((256, 256, 3), dtype=np.uint8)
mask = np.zeros((256, 256), dtype=np.uint8)
rr, cc = disk((128, 128), 60)
mask[rr, cc] = 1
image[rr, cc] = [180, 30, 30]  # 模拟病灶

plt.figure(figsize=(6, 3))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Toy Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(mask, cmap='gray')
plt.title('Toy Mask')
plt.axis('off')
plt.show()
````

## 2. 下载 SAM 权重

**做什么**：下载预训练模型权重。
**为什么**：SAM 不是从零训练的，我们通常使用官方权重做推理或微调。
**结果**：得到 `sam_vit_b_01ec64.pth` 文件。


````python
import os
import urllib.request

os.makedirs('data/sam', exist_ok=True)
weight_path = 'data/sam/sam_vit_b_01ec64.pth'
if not os.path.exists(weight_path):
    url = 'https://dl.fbaipublicfiles.com/segment_anything/sam_vit_b_01ec64.pth'
    urllib.request.urlretrieve(url, weight_path)
    print('Downloaded SAM weights.')
else:
    print('SAM weights already exist.')
````

## 3. 使用 SAM 进行推理

**做什么**：输入一张图像 + 一个前景点，得到分割结果。
**为什么**：SAM 以“提示点/框”引导模型生成掩码。
**结果**：得到预测分割掩码。


````python
import torch
from segment_anything import sam_model_registry, SamPredictor

device = 'cuda' if torch.cuda.is_available() else 'cpu'
sam = sam_model_registry['vit_b'](checkpoint=weight_path)
sam.to(device=device)
predictor = SamPredictor(sam)

predictor.set_image(image)
# 提示点：我们给出病灶中心点 (128, 128)
input_point = np.array([[128, 128]])
input_label = np.array([1])  # 1 表示前景

masks, scores, logits = predictor.predict(
    point_coords=input_point,
    point_labels=input_label,
    multimask_output=True
)

best_mask = masks[np.argmax(scores)]

plt.figure(figsize=(6, 3))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.scatter([128], [128], c='yellow', s=40)
plt.title('Input + Prompt')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(best_mask, cmap='gray')
plt.title('SAM Output')
plt.axis('off')
plt.show()
````

## 4. 评价指标：Dice & IoU

**做什么**：计算模型分割的好坏。
**为什么**：医学图像分割常用 Dice/IoU 衡量重叠程度。
**结果**：打印数值。


````python
def dice_coef(pred, target, eps=1e-6):
    pred = pred.astype(np.float32)
    target = target.astype(np.float32)
    intersection = (pred * target).sum()
    return (2.0 * intersection + eps) / (pred.sum() + target.sum() + eps)

def iou_score(pred, target, eps=1e-6):
    pred = pred.astype(np.float32)
    target = target.astype(np.float32)
    intersection = (pred * target).sum()
    union = pred.sum() + target.sum() - intersection
    return (intersection + eps) / (union + eps)

dice = dice_coef(best_mask, mask)
iou = iou_score(best_mask, mask)
print(f'Dice: {dice:.4f}, IoU: {iou:.4f}')
````

## 5. （可选）轻量微调思路

**做什么**：冻结大部分参数，仅训练 Mask Decoder。
**为什么**：SAM 很大，完整训练代价高，常用轻量微调。
**结果**：你可以在自己的数据上提升分割效果。

下面给出“可运行的最小代码结构”，你只需要替换数据加载部分即可开始训练。

````python
import torch
from torch.utils.data import Dataset, DataLoader

class ToySegDataset(Dataset):
    def __len__(self):
        return 8

    def __getitem__(self, idx):
        return image, mask

dataset = ToySegDataset()
loader = DataLoader(dataset, batch_size=1, shuffle=True)

# 冻结图像编码器，只训练 mask decoder
for name, param in sam.image_encoder.named_parameters():
    param.requires_grad = False

optimizer = torch.optim.Adam(sam.mask_decoder.parameters(), lr=1e-4)
loss_fn = torch.nn.BCEWithLogitsLoss()

sam.train()
for epoch in range(1):
    for batch_img, batch_mask in loader:
        predictor.set_image(batch_img[0].numpy())
        masks, scores, logits = predictor.predict(
            point_coords=input_point,
            point_labels=input_label,
            multimask_output=False
        )
        pred = torch.tensor(masks, dtype=torch.float32)
        gt = torch.tensor(batch_mask.numpy(), dtype=torch.float32)
        loss = loss_fn(pred, gt)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        print(f'Epoch {epoch} Loss: {loss.item():.4f}')
````
